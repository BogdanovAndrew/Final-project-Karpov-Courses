1. В игре Plants & Gardens каждый месяц проводятся тематические события, ограниченные по времени. В них игроки могут получить уникальные предметы для сада и персонажей, дополнительные монеты или бонусы. Для получения награды требуется пройти ряд уровней за определенное время. С помощью каких метрик можно оценить результаты последнего прошедшего события?

Для ответа на поставленный вопрос рассмотрим следующие продуктовые метрики по категориям: 

# Аудитория 

**Средняя дневная аудитория** — среднее арифметическое DAU за период проведения события. Покажет, насколько изменился онлайн в игре по сравнению с обычными показателями вне события/предыдущими событиями.

**Количество отвалившихся/вернувшихся пользователей** — число новых пользователей за данный период времени. Может показать, насколько эффективно сработал пиар события и насколько оно успешно. Во многих играх это в основном показывается на количестве именно вернувшихся игроков.

**Доля новых пользователей** — отношение числа новых пользователей ко всем пользователям за этот же промежуток времени. Было ли наше событие настолько раскрученным, что смогло присоединить к продукту новых игроков и потенциальных клиентов.

**Среднее время, проведенное в игре** - должно увеличиться в случае успеха ивента и держаться выше планки внесобытийного времени в игре.

# Воронка, конверсия, удержание

**Воронка** — последовательность действий пользователей до совершения ключевого действия, в нашем случае до попадания на каждый конкретный уровень. Показывает, как пользователи проходят через каждый уровень в событии, на каких шагах и какая часть из них отваливается.

**Конверсия (CR)** — отношение числа пользователей, которые прошли уровень и получили награду к общему числу пользователей, которые дошли до этого уровня (метрика связана с воронкой). Наиболее точно будет оценивать относительную конверсию по отношению от уровня к уровню.

**Retention (RR)** - метрика, которая показывает, как пользователи возвращаются в продукт (и часто свидетельствует, нравится ли он им). Соответственно в период успешного события этот показатель (процент пользователей) должен быть выше обычного.

Данные метрики помогут оценить правильность поставленных условий прохождения каждого уровня, наличия ошибок как на программном, так и внутриигровых механиках, понять успешность события и сформировать план действий на текущий этап.

# Показатели доходов

**Средний чек (AOV)** - это доход, разделённый на количество покупок. Поскольку в подобных уровнях часть или все уровни могут покупаться за внутриигровую валюту, получаемую путем доната, и в случае, если событие интересное и награда за прохождение уровней и/или финальная являются ценными и достойными, то разумеется данная метрика должна показать рост.

**ARPPU (Average revenue per paying user)** - средний доход на платящего пользователя (то есть клиента). Состоит из того, насколько часто пользователь покупает, и насколько большой средний чек покупки. Зависит от AOV, но не учитывает бездонатных пользователей. Принцип работы и выводов по аналогии описанной в среднем чеке.

# Метрики лояльности и внутриигровые показатели

**NPS (net promoter score)** — метрика уровня лояльности пользователей. Измеряется с помощью опросников, отражается в процентах. Можно спрашивать у пользователей, нравится ли им текущее событие, удовлетворены ли они наградой и тому подобнее, дать возможность дать оценку от 1 до 10 и делить их на промоутеров и детракторов. Также дополнить ее сегментацией пользователей по группам, тем самым получим напрямую оценку нашего ивента. Можно заменить также на CSAT.

Также важно подсчитывать и анализировать внутриигровые достижения. Сделать это можно основываясь на следующих показателях:
1. Среднее количество заработанных предметов, монет и бонусов 
2. Число игроков, завершивших ивент
3. Сегментация пользователей по количеству заработанных достижений, размеры сегментов.
4. Показатель проходимость каждого уровня и подсчитать среднее время прохождения каждого уровня.
5. Сравнение число участников ивента и повторных участников с предыдущими событиями.

Все вышеописанные показатели и метрики позволят сформировать четкое понимание об успешности, прибыльности и развитии продукта на период проведения ивента. Также позволят сравнивать полученные данные с данными других событий с помощью статистических тестов. Также позволят проводить дополнительные A/B тесты для правильной интерпретации результатов.

2. Предположим, в другом событии мы усложнили механику событий так, что при каждой неудачной попытке выполнения уровня игрок будет откатываться на несколько уровней назад. Изменится ли набор метрик оценки результата? Если да, то как?

Как правило при добавлении чего-то нового - нового функционала, дизайна, МЕХАНИКИ СОБЫТИЯ - важно посмотреть, как отреагировали наши пользователи и клиенты. Для этого нужно измерить все значимые показатели, например описываемые ранее, и с помощью статистических тестов посмотреть значимо ли изменились наши метрики и главное в какую сторону. Поэтому большинство метрик я бы измерял повторно, присваивал бы статус тестовой группы и сравнивал с предыдущим - контрольной. Проводил бы своего рода A/B тест, но уже на версии продакшена.

Однако с учетом появления новых механик, также стоит учитывать некоторые показатели и метрики, которых не было до этого. Я бы выделил следующие:

# Дополнительные показатели

1. Churn rate - коэффициент оттока клиентов. Метрика по сути обратная RR, однако она при этих этапах является более главной, поскольку важно понимать, не нанесут ли наши изменения ущерб экономике продукта и его доходности. Может измеряться в других временных интервалах, не таких как RR.
2. Количество перепрохождений: эта новая метрика покажет, сколько раз в среднем у игроков сбрасывался их прогресс при прохождении события. Это поможет оценить, насколько сложной была новая механика.
3. Количество повторных попыток: эта новая метрика покажет, сколько раз в среднем игроки повторяли один и тот же уровень, прежде чем пройти его.
4. Еще один пункт отведу на анализ таких важных показателей, как NPS, ARPPU и полученные внутриигровые ресурсы. За этим стоит пристально наблюдать, чтобы в случае чего не терять целевую аудиторию и клиентов.

В целом это все важные показатели, которые бы я учитывал, будучи продуктовым аналитиком данной игры. Они помогут сложить полную картину происходящего и вовремя и правильно реагировать на появляющиеся изменения.
